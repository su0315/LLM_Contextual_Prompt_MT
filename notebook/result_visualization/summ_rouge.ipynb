```python
def get_and_return_scores(scenario, model):
    rouge1_list, rouge2_list, rougel_list, rougelsum_list, lang_list = get_score(scenario, model)
    scores = {
        'model': model,
        'scenario': scenario,
        'rouge1_scores': rouge1_list,
        'rouge2_scores': rouge2_list,
        'rougeL_scores':rougel_list,
        'rougelsum_scores':rougelsum_list
    }
    return scores, lang_list

"""
def get_and_return_scores_b(scenario, model):
    comet_list_b, bleu_list_b, lang_list_b = get_score_b(scenario, model)
    scores_b = {
        'model': model,
        'scenario': scenario,
        'comet_scores': comet_list_b,
        'bleu_scores': bleu_list_b,
    }
    return scores_b, lang_list_b
"""

models = ["transforemersum-distilbert-ctpro"]
scenarios = ["1-1", "2-1", "3-1", "4-1", "5-1", "6-1", "7-1", "2-1-ante", "ante-1"]

results = []


for model in models:
    for scenario in scenarios:
        #print (scenario)
        scores, lang_list = get_and_return_scores(scenario, model)
        results.append(scores)


# Now you have a list of dictionaries containing the scores for each model and scenario
# You can access the results as needed
for result in results:
    print(f"Model: {result['model']}, Scenario: {result['scenario']}")
    print(f"Rouge1 Scores: {result['rouge1_scores']}")
    print(f"Rouge1 Scores: {result['rouge2_scores']}")
    #print(f"ROUGEL Scores: {result['rougel_scores']}")
    #print(f"Rouge1 Scores: {result['rougelsum_scores']}")
    print()
```
